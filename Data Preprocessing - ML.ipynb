```python
# Melt the data to reshape it
df_melted = df.melt(id_vars=['Country Name', 'Country Code', 'Indicator Name'], var_name='Year', value_name='Value')

# Filter out rows where Year is not numeric
df_melted = df_melted[pd.to_numeric(df_melted['Year'], errors='coerce').notnull()]

# Pivot the data to have Indicator Name as columns
df_pivoted = df_melted.pivot(index=['Country Name', 'Country Code', 'Year'], columns='Indicator Name', values='Value').reset_index()

# Rename the index name
df_pivoted.index.name = None

df_pivoted = df_pivoted.drop(df_pivoted.columns[:3], axis=1)

df_pivoted.head()
```

```python
missing_percentage = (df_pivoted.isnull().mean() * 100).round(2)

print("Percentage of missing values in each column:")
print(missing_percentage)
```

```python
threshold = 30
cols_to_drop = missing_percentage[missing_percentage > threshold].index
df_dropped = df_pivoted.drop(columns=cols_to_drop)
print("\nColumns to drop due to high percentage of missing values:")
print(cols_to_drop)

print("\nDataFrame after dropping columns:")
```

```python
missing_checked_percentage = (df_dropped.isnull().mean() * 100).round(2)

print("Percentage of missing values in each column:")
print(missing_checked_percentage)
```

```python
import pandas as pd
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor

# Initialize the IterativeImputer
rf_imputer = IterativeImputer(estimator=RandomForestRegressor(), max_iter=10, random_state=42)


imputed_data = rf_imputer.fit_transform(df_dropped)
imputed_df = pd.DataFrame(imputed_data, columns=df_dropped.columns)
```

```python
imputed_df.head()
```

```python
import numpy as np
import pandas as pd
from scipy import stats

def remove_outliers_zscore(df, columns, threshold=3):
    df_clean = df.copy()
    for column in columns:
        z_scores = np.abs(stats.zscore(df[column]))
        df_clean = df_clean[z_scores <= threshold]
    return df_clean

def remove_outliers_iqr(df, columns, factor=1.5):
    df_clean = df.copy()
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - factor * IQR
        upper_bound = Q3 + factor * IQR
        df_clean = df_clean[(df_clean[column] >= lower_bound) & (df_clean[column] <= upper_bound)]
    return df_clean

def remove_outliers_mad(df, columns, threshold=3.5):
    df_clean = df.copy()
    for column in columns:
        median = df[column].median()
        mad = np.median(np.abs(df[column] - median))
        lower_bound = median - threshold * mad
        upper_bound = median + threshold * mad
        df_clean = df_clean[(df_clean[column] >= lower_bound) & (df_clean[column] <= upper_bound)]
    return df_clean

# Remove outliers using all three methods
df_cleaned_zscore = remove_outliers_zscore(imputed_df, imputed_df.columns)
df_cleaned_iqr = remove_outliers_iqr(imputed_df, imputed_df.columns)
df_cleaned_mad = remove_outliers_mad(imputed_df, imputed_df.columns)

def print_results(method_name, original_df, cleaned_df):
    rows_removed = len(original_df) - len(cleaned_df)
    percentage_removed = (rows_removed / len(original_df)) * 100
    print(f"\n{method_name} Method:")
    print(f'Rows in original dataset: {len(original_df)}')
    print(f'Rows in cleaned dataset: {len(cleaned_df)}')
    print(f'Number of rows removed: {rows_removed}')
    print(f'Percentage of rows removed: {percentage_removed:.2f}%')

    
print_results("Z-Score", imputed_df, df_cleaned_zscore)
print_results("IQR", imputed_df, df_cleaned_iqr)
print_results("MAD", imputed_df, df_cleaned_mad)
```

```python
import numpy as np
import pandas as pd
from scipy import stats

z_scores = np.abs(stats.zscore(imputed_df))

# Set threshold = 5
threshold = 5 
df_cleaned_zscore = imputed_df[(z_scores < threshold).all(axis=1)]

rows_original = len(imputed_df)
rows_cleaned = len(df_cleaned_zscore)
rows_removed = rows_original - rows_cleaned
percentage_removed = (rows_removed / rows_original) * 100

print("\nZ-Score Method:")
print(f'Rows in original dataset: {rows_original}')
print(f'Rows in cleaned dataset: {rows_cleaned}')
print(f'Number of rows removed: {rows_removed}')
print(f'Percentage of rows removed: {percentage_removed:.2f}%')
```
