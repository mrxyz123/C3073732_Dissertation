```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = 'Primary_DataSet_Raw.csv'

primary_df = pd.read_csv(file_path)
```

```python
primary_df.head()
```

```python
secondary_file_path = 'Country_Abbr.csv'

# Read the secondary CSV file
secondary_df = pd.read_csv(secondary_file_path)
secondary_df.head()
```

```python
# Perform the anti join to find records in df1 that do not have a match in df2
df_not_country = primary_df[~primary_df['Country Code'].isin(secondary_df['alpha-3'])]
df_not_country.head()
```

```python
# Find the unique unqualified country or organisation 
df_not_country['Country Name'].unique()
```

```python
# Find the common keys
common_keys = primary_df['Country Code'].isin(secondary_df['alpha-3'])

# Keep only the records from df1 that have a match in df2
df = primary_df[common_keys]
df.head()
```

```python
df = pd.DataFrame(df)

# Select columns from 2 to 5 (inclusive) and 9 to 11 (inclusive)
final_df = pd.concat([df.iloc[:, 0:4], df.iloc[:, -25:]], axis=1)

final_df.head()
```

```python
final_df.to_csv('final_df.csv', index=False)
```
