```python
import matplotlib.pyplot as plt

ratings = ['1 (Excellent)', '2 (Very Good)', '3 (Average)', '4 (Fair)', '5 (Poor)']
counts = [3, 7, 5, 0, 0]


plt.figure(figsize=(8, 5))
plt.bar(ratings, counts, color='skyblue')
plt.title("User Ratings for the App's Visual Design")
plt.xlabel('Rating')
plt.ylabel('Number of Responses')

for i, count in enumerate(counts):
    plt.text(i, count + 0.1, str(count), ha='center', va='bottom')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
ratings = [1, 2, 3, 4, 5]
counts = [3, 7, 5, 0, 0]

total_responses = sum(counts)
percentages = [(count / total_responses) * 100 for count in counts]

mean = sum(rating * count for rating, count in zip(ratings, counts)) / total_responses
median = ratings[1]  
mode = ratings[counts.index(max(counts))]


print("\nPercentage Distribution:")
for rating, percentage in zip(ratings, percentages):
    print(f"Rating {rating}: {percentage:.2f}%")

print("\nCentral Tendency:")
print(f"Mean: {mean:.2f}")
print(f"Median: {median}")
print(f"Mode: {mode}")
```
